```python
from train import train
import config
```

```python
target_update_freqs = [10, 100, 1000]
returns_list = []

for target_update_freq in target_update_freqs:
    conf = config.CartPole
    conf['target_update_frequency'] = target_update_freq
    returns = train(config=conf)
    returns_list.append(returns)
```

```python
# Create plots
import matplotlib.pyplot as plt

# Data
episodes = [i for i in range(1, 100, 25)]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(episodes, returns_list[0], marker='o', linestyle='-', color='b', label='target_update_frequency=10')
plt.plot(episodes, returns_list[1], marker='o', linestyle='-', color='r', label='target_update_frequency=100')
plt.plot(episodes, returns_list[2], marker='o', linestyle='-', color='g', label='target_update_frequency=1000')
plt.title('Policy Evaluation over 500 Episodes')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('target_update_frequency.png')
```
